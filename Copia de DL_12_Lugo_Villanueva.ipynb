```python
#Función general de plt.imshow
#La imagen está en blanco y negro
def plt_imshow(img, title:str='',show:bool=True, return_img:bool=False):
    img = cv.cvtColor(img, cv.COLOR_BGR2RGB)    
    plt.imshow(img)
    plt.title(title)
    if show: plt.show()
    if return_img: return img
```

```python
#img es la imagen a analizar
#model es el modelo que se usará para detectar el estado de ánimo
#face es una tupla que contiene los parámetros para detectar los rostros em una imagen
#imgTitle y yTitle contienen los títulos de la imágenes
def face_recon(img,model,face:tuple=(1.1,15),
               imgTitle:str='',yTitle:int=0.75):
    #img_original = img.copy()
    img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

    detector_rostros = cv.CascadeClassifier('DL_12_UADY/haarcascade_frontalface_default.xml')
    faces = detector_rostros.detectMultiScale(img, face[0],face[1])

    for (x,y,w,h) in faces:
        cv.rectangle( img , (x,y),(x+w, y+h), (0,255,0),8 )
        img_face = img_gray[x:x+w,y:y+h]
        try:
          img_face=cv.resize(img_face,(40,40),interpolation = cv.INTER_NEAREST)
          state = model.predict(img_face.reshape((1,1600)))[0]
          if state == 0:
            cv.putText(img,text='Negative :(',org=(x,y+h),fontFace=cv.FONT_HERSHEY_SIMPLEX,fontScale=2,color=(255,255,0),thickness=3, lineType=cv.LINE_AA)
          elif state == 1:
            cv.putText(img,text='Positive :)',org=(x,y+h),fontFace=cv.FONT_HERSHEY_SIMPLEX,fontScale=2,color=(0,255,255),thickness=3, lineType=cv.LINE_AA)
        except: pass
    
    plt_imshow(img)
```
